```python
attributes = BP1_df.columns[:-1]  # Exclude 'U1_US' column

# Create a dictionary to store cross-tabulations
cross_tabs = {}

# Iterate over attributes and perform cross-tabulation with 'U1_US'
for attribute in attributes:
    cross_tab = pd.crosstab(index=BP1_df['U1_US'], columns=BP1_df[attribute])
    cross_tabs[attribute] = cross_tab

# Display the cross-tabulations
for attribute, cross_tab in cross_tabs.items():
    print(f"Cross-tabulation for {attribute}:")
    print(cross_tab)
    print()
```

```python
# Select attributes columns (excluding 'U1_US')
attributes = BP1_df.columns[:-1]

# Create a new DataFrame for the desired output
output_data = []

# Iterate over each attribute
for attribute in attributes:
    # Calculate the sum of occurrences of value 1 for each 'U1_US' group
    sum_values = BP1_df.groupby('U1_US')[attribute].apply(lambda x: (x == 1).sum())
    
    # Append the sums to the output_data list
    output_data.append(sum_values)

# Create the output DataFrame
output_df = pd.concat(output_data, axis=1)
#output_df.columns = ['BP1', 'BP2', 'BP3', 'BP4', 'BP5', 'BP6', 'BP7', 'BP8', 'BP9', 'BP10', 'BP11', 'BP12', 'BP13', 'BP14', 'BP15', 'BP16', 'BP17', 'BP97', 'BP98']
output_df.columns=attributes
# Display the output DataFrame
output_df
```
